```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Real estate.csv')
```

```python
df.head()
```

```python
df = df.drop(['No','X5 latitude','X6 longitude'], axis=1)
```

```python
df['X1 transaction date'] = df['X1 transaction date'].astype(str).str.extract(r'^(\d{4})')[0]
```

```python
df.head()
```

# Split the Dataset

```python
X = df[['X1 transaction date', 'X2 house age', 'X3 distance to the nearest MRT station', 'X4 number of convenience stores']]
Y = df['Y house price of unit area']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)
```

```python
x_test
```

```python
y_test
```

```python
df.corr()
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
LR = LinearRegression()
```

```python
df.head()
```

```python
LR.fit(x_train, y_train)
```

```python
LR.score(x_train, y_train)
```

```python
y_predict_linear = LR.predict(x_test)
```

```python
y_predict_linear[:5]
```

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=3)
X_train_poly = poly.fit_transform(x_train)
X_test_poly = poly.transform(x_test)
```

```python
X_train_poly.shape
```

```python
pd.DataFrame(X_poly)
```

```python
poly_model = LinearRegression()
```

```python
poly_model.fit(X_train_poly, y_train)
```

```python
Y_pred_poly = poly_model.predict(X_test_poly)
```

```python
Y_pred_poly[:5]
```

# Evaluate Model Performance

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, Y_predict_linear)
r2= r2_score(y_test, Y_predict_linear)
```

```python
mse
```

```python
r2
```

#  Actual vs. Predicted Values for Both Models

```python
import matplotlib.pyplot as plt

# Plot actual vs predicted for Linear Regression
plt.figure(figsize=(12, 6))
plt.scatter(y_test, Y_predict_linear, color='blue', label='Linear Regression')
plt.scatter(y_test, Y_pred_poly, color='red', label='Polynomial Regression')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='green', label='Perfect Fit')

plt.title('Actual vs. Predicted Values')
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.legend()
plt.show()
```

# Curve Fitting for Polynomial Regression

```python
import numpy as np

# Sort data for smooth plotting
sorted_idx = np.argsort(x_test['X2 house age'])  # Example: Sorting by house age
X_test_sorted = x_test.iloc[sorted_idx]
Y_test_sorted = y_test.iloc[sorted_idx]
Y_pred_poly_sorted = Y_pred_poly[sorted_idx]

plt.figure(figsize=(12, 6))
plt.plot(X_test_sorted['X2 house age'], Y_test_sorted, 'o', label='Actual Values')
plt.plot(X_test_sorted['X2 house age'], Y_pred_poly_sorted, '-', label='Polynomial Fit', color='red')

plt.title('Polynomial Regression Curve Fit')
plt.xlabel('X2 (House Age)')
plt.ylabel('Y (House Price per Unit Area)')
plt.legend()
plt.show()
```
